# Basic Recommendation System

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns; sns.set_style('white')

import warnings; warnings.filterwarnings('ignore')

%matplotlib inline
```

## Loading Data Set and Data Exploration

```python
columns_names = ['user_id', 'item_id', 'rating', 'timestamp']
data = pd.read_csv('./datasets/recom/u.data', sep='\t', names=columns_names)
```

```python
data.head()
```

```python
movie_titles = pd.read_csv('./datasets/recom/movie_id_titles.csv')
movie_titles.head()
```

```python
data = pd.merge(data, movie_titles, on='item_id')
data.head()
```

```python
data.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
data.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(data.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['num of ratings'] = pd.DataFrame(data.groupby('title')['rating'].count())
ratings.sort_values('num of ratings', ascending=False).head(10)
```

```python
ratings['num of ratings'].hist(bins = 70)
```

```python
ratings['rating'].hist(bins = 70)
```

```python
sns.jointplot(x='rating', y='num of ratings', data=ratings, alpha=0.5)
```

```python
movie_mat = data.pivot_table(index='user_id', columns='title', values='rating')
movie_mat.head(10)
```

## Similar Movies to Star Wards

```python
starwars_user_ratings = movie_mat['Star Wars (1977)']
starwars_user_ratings.head()
```

```python
starwars_corr = movie_mat.corrwith(starwars_user_ratings)
starwars_corr.head()
```

```python
starwars_corr = pd.DataFrame(starwars_corr, columns=['Correlation']).dropna()
starwars_corr.sort_values('Correlation', ascending=False).head(10)
```

```python
starwars_corr = starwars_corr.join(ratings['num of ratings'])
starwars_corr.head()
```

```python
starwars_corr[starwars_corr['num of ratings'] > 100].sort_values('Correlation', ascending=False).head()
```
